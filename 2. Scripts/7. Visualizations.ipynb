# 1

```python
# Create barchart of orders for days of the week
barchart = dataframe['order_day_of_week'].value_counts().sort_index().plot.bar(color = ['orange', 'yellow'])
```

```python
# Export barchart
barchart.figure.savefig(r"C:\Users\justi\Downloads\Instacart Basket Analysis\4. Analysis\Visualizations\barchart.png")
```

```python
# Create histogram of product prices
histogram = dataframe['prices'].plot.hist(bins = 50)
```

```python
# Export histogram
histogram.figure.savefig(r"C:\Users\justi\Downloads\Instacart Basket Analysis\4. Analysis\Visualizations\histogram.png")
```

```python
# Create scatterplot of prices to check for irregularities
seaborn.scatterplot(x = 'prices', y = 'prices', data = dataframe)
```
